## ДџЕёВ§В© 2. MODEL

```python
from unsloth import FastLanguageModel
import torch

max_seq_length = 2048

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/llama-3-8b-instruct-bnb-4bit",
    max_seq_length = max_seq_length,
    load_in_4bit = True,
)

model = FastLanguageModel.get_peft_model(
    model,
    r = 32,
    target_modules = [
        "q_proj","k_proj","v_proj","o_proj",
        "gate_proj","up_proj","down_proj",
    ],
    lora_alpha = 32,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
)
```

## ДџЕёВ§В© 3. DATASET

```python
from datasets import load_dataset

magicoder = load_dataset(
    "ise-uiuc/Magicoder-OSS-Instruct-75K",
    split="train"
)

raw_code = load_dataset(
    "bigcode/the-stack-v2",
    split="train",
    streaming=True
)
```

## ДџЕёВ§В© 4. ДџЕёвЂќВҐ MODERN FILTER

```python
MODERN_KEYWORDS = [
    # Android
    "androidx.compose","compose.material","compose.ui",
    "viewmodel","stateflow","sharedflow", "kotlin",
    "dagger.hilt","@inject",
    "kotlinx.coroutines","suspend fun","flow<",
    "retrofit2","okhttp3","ktor",
    "room","coil","navigation-compose",

    # Backend
    "springframework.boot","spring web","spring data",
    "fastapi","pydantic","sqlalchemy","uvicorn", "java",
    "python",

    # JS
    "express","nestjs","prisma"
]

def is_good_code(example):
    code = example["content"].lower()
    lang = example["language"]

    if lang not in ["Kotlin","Java","Python","JavaScript"]:
        return False

    score = sum(1 for kw in MODERN_KEYWORDS if kw in code)

    return (
        score >= 3 and
        len(code.splitlines()) > 10 and
        "test" not in code
    )
```

## ДџЕёВ§В© 5. ДџЕёвЂќВҐ INSTRUCTION GENERATION

```python
import re

def create_instruction(code):
    c = code.lower()
    # Kodun iГѓВ§inden sГ„В±nГ„В±f veya fonksiyon ismini ГѓВ§ekmeye ГѓВ§alГ„В±Г…ЕёalГ„В±m (ГѓвЂ“rn: class LoginScreen -> LoginScreen)
    name_match = re.search(r'(?:class|fun|def|const)\s+([a-zA-Z0-9_]+)', code)
    name = name_match.group(1) if name_match else "this component"

    if "compose" in c:
        return f"Implement a modern Jetpack Compose UI for '{name}' using Material 3 and best practices."
    elif "viewmodel" in c:
        return f"Create an Android ViewModel for '{name}' using StateFlow and clean architecture principles."
    elif "retrofit" in c:
        return f"Define a Retrofit API service interface for '{name}' with proper error handling."
    elif "spring" in c:
        return f"Build a high-performance Spring Boot REST Controller/Service for '{name}'."
    elif "fastapi" in c:
        return f"Develop a production-ready FastAPI endpoint for '{name}' with Pydantic validation."
    elif "@inject" in c or "hilt" in c:
        return f"Setup Dagger-Hilt dependency injection for '{name}' in an Android context."

    return f"Analyze, explain, and refactor the following code for better performance: {name}"
```

## ДџЕёВ§В© 6. RAW ГўвЂ вЂ™ DATASET

```python
from tqdm.auto import tqdm

filtered_samples = []
processed = 0
seen = set()

# HГ„В±z ve durumu canlГ„В± gГѓВ¶rmek iГѓВ§in progress bar ekliyoruz
pbar = tqdm(desc="Г„В°ncelenen Dosya", unit=" dosya")

for example in raw_code:
    processed += 1
    pbar.update(1) # Her dosyada barГ„В± gГѓВјnceller

    lang = example.get("language")

    # ДџЕёЕЎв‚¬ HIZLANDIRMA 1: EГ„Еёer dil bizim aradГ„В±klarГ„В±mГ„В±zdan deГ„Еёilse,
    # kodu indirmeye/okumaya bile ГѓВ§alГ„В±Г…Еёma, direkt atla.
    if lang not in ["Kotlin", "Java", "Python", "JavaScript"]:
        continue

    code = example.get("content")

    # ДџЕёвЂќВґ gГѓВјvenlik
    if not code or not isinstance(code, str):
        continue

    code = code.strip()

    # ДџЕёвЂќВґ ГѓВ§ГѓВ¶p filtre
    if len(code) < 200:
        continue

    # ДџЕёвЂќВґ gerГѓВ§ek kod kontrolГѓВј
    if "import" not in code and "class" not in code and "fun" not in code:
        continue

    # ДџЕёвЂќВґ modern filter (Г„В°ГѓВ§inde "compose", "spring" vb. geГѓВ§iyor mu?)
    if not is_good_code(example):
        continue

    # ДџЕёвЂќВґ duplicate engelle
    # Hash iГ…Еёlemini tГѓВјm testleri geГѓВ§en kodlara saklГ„В±yoruz ki CPU yorulmasГ„В±n
    code_hash = hash(code[:500])
    if code_hash in seen:
        continue
    seen.add(code_hash)

    # ДџЕёвЂќВґ truncate
    code = code[:2000]

    instruction = create_instruction(code)

    if not instruction or len(instruction) < 10:
        continue

    filtered_samples.append({
        "instruction": instruction,
        "output": code,
        "context": "Modern Android and backend development best practices."
    })

    # Г„В°lerlemeyi Progress Bar'Г„В±n yanГ„В±na canlГ„В± olarak yazdГ„В±r
    if processed % 100 == 0:
        pbar.set_postfix({"SeГѓВ§ilen": len(filtered_samples)})

    if len(filtered_samples) >= 8000:
        break

pbar.close()
print(f"ГўЕ“вЂ¦ Г„В°Г…Еёlem tamam! Toplam {processed} dosya tarandГ„В±, {len(filtered_samples)} ГѓВ¶rnek baГ…ЕёarГ„В±yla seГѓВ§ildi.")
```

## ДџЕёВ§В© 7. FORMAT

```python
# Llama-3 Instruct benzeri ama Alpaca yapГ„В±sГ„В±nda stabilize edilmiГ…Её format
train_prompt_style = """<|begin_of_text|><|start_header_id|>system<|end_header_id|>

You are a senior Android (Kotlin/Compose) and Backend (Java/Python) engineer. Provide clean, modern, and efficient code.<|eot_id|><|start_header_id|>user<|end_header_id|>

Context: {context}
Instruction: {instruction}<|eot_id|><|start_header_id|>assistant<|end_header_id|>

{output}<|eot_id|>"""

def format_data(data):
    out = []
    for x in data:
        instruction = x.get("instruction", "")
        output = x.get("output") or x.get("response")  # ДџЕёвЂќВҐ FIX

        if not instruction or not output:
            continue

        context = x.get("context", "General best practices for modern software development.")

        text = train_prompt_style.format(
            context=context,
            instruction=instruction,
            output=output
        )

        text += tokenizer.eos_token
        out.append({"text": text})

    return out

magicoder_data = format_data(magicoder)
raw_data = format_data(filtered_samples)
```

## ДџЕёВ§В© 8. MERGE

```python
import random
from datasets import Dataset

final_data = magicoder_data[:20000] + raw_data
random.shuffle(final_data)

train_dataset = Dataset.from_list(final_data)
```

## ДџЕёВ§В© 9. TRAIN

```python
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model=model,
    train_dataset=train_dataset,
    dataset_text_field="text",
    max_seq_length=2048,
    args=TrainingArguments(
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4, # T4 GPU iГѓВ§in 4-8 arasГ„В± idealdir
        warmup_steps=10,
        #max_steps=500, # 20k veri iГѓВ§in en az 500-1000 arasГ„В± ГѓВ¶neririm
        num_train_epochs=1
        learning_rate=5e-5, # Daha kararlГ„В± bir ГѓВ¶Г„Еёrenme iГѓВ§in hafif dГѓВјГ…ЕёГѓВјrdГѓВјk
        fp16 = not torch.cuda.is_bf16_supported(),
        bf16 = torch.cuda.is_bf16_supported(),
        logging_steps=10,
        save_steps=100, # Her 100 adГ„В±mda bir modeli kaydet (Colab ГѓВ§ГѓВ¶kerse hayat kurtarГ„В±r)
        output_dir="outputs",
        optim = "adamw_8bit", # Bellek tasarrufu
    ),
)

trainer.train()
```

## ДџЕёВ§В© 11. ДџЕёвЂќВҐ RAG SETUP

```python
from sentence_transformers import SentenceTransformer
import chromadb

embed_model = SentenceTransformer("all-MiniLM-L6-v2")

client = chromadb.Client()
collection = client.create_collection("docs")
```

## ДџЕёВ§В© 12. DOC EKLE

```python
import json

file_path = "/content/drive/MyDrive/Colab Notebooks/data.json"

with open(file_path, "r", encoding="utf-8") as f:
    json_data = json.load(f)

print(f"Toplam {len(json_data)} dokГѓВјman yГѓВјklendi")
```

```python
def chunk_text(text, chunk_size=500, overlap=100):
    words = text.split()
    chunks = []

    for i in range(0, len(words), chunk_size - overlap):
        chunk = " ".join(words[i:i+chunk_size])
        chunks.append(chunk)

    return chunks
```

```python
def process_json(doc):
    chunks = []

    # ДџЕёвЂќВ№ TEXT chunk
    content = doc.get("content", "")
    text_parts = chunk_text(content, 500)

    for part in text_parts:
        chunks.append({
            "text": part,
            "category": doc.get("category", ""),
            "topic": doc.get("topic", ""),
            "source": doc.get("source", ""),
            "type": "text"
        })

    # ДџЕёвЂќВ№ CODE chunk
    codes = doc.get("code_snippets") or [doc.get("code", "")]
    for code in codes:
        if code and len(code) > 15:
            chunks.append({
                "text": code,
                "category": doc.get("category", ""),
                "topic": doc.get("topic", ""),
                "source": doc.get("source", ""),
                "type": "code"
            })

    return chunks
```

```python
def add_json_to_rag(json_data):
    all_chunks = []
    for doc in json_data:
        chunks = process_json(doc)
        all_chunks.extend(chunks)

    print(f"Toplam chunk sayГ„В±sГ„В±: {len(all_chunks)}")

    # Embeddingleri topluca al (ГѓвЂЎok daha hГ„В±zlГ„В±dГ„В±r)
    texts = [chunk["text"] for chunk in all_chunks]
    embeddings = embed_model.encode(texts).tolist()

    metadatas = [{
        "category": chunk["category"],
        "type": chunk["type"],
        "topic": chunk["topic"],
        "source": chunk["source"]
    } for chunk in all_chunks]

    ids = [f"doc_{i}" for i in range(len(all_chunks))]

    # VeritabanГ„В±na topluca ekle
    collection.add(
        documents=texts,
        embeddings=embeddings,
        metadatas=metadatas,
        ids=ids
    )

    print("ГўЕ“вЂ¦ TГѓВјm veriler RAG'e eklendi")
```

```python
add_json_to_rag(json_data)
```

## ДџЕёВ§В© 13. RETRIEVE

```python
def retrieve(query, category=None):
    emb = embed_model.encode(query).tolist()

    where = {}

    if category:
        where["category"] = category

    results = collection.query(
        query_embeddings=[emb],
        n_results=7,
        where=where if where else None
    )

    return results["documents"][0]
```

## ДџЕёВ§В© 14. ДџЕёвЂќВҐ CLAUDE-LIKE PROMPT

```python
def build_prompt(user_query):
    docs = retrieve(user_query)
    context = "\n".join(docs) if docs else "No specific documentation found."

    return f"""<|begin_of_text|><|start_header_id|>system<|end_header_id|>

You are a senior Android (Kotlin/Compose) and Backend (Java/Python) engineer.

Rules:
- Always use modern best practices
- Prefer Jetpack Compose
- Use clean architecture
- Explain briefly, then provide code<|eot_id|><|start_header_id|>user<|end_header_id|>

Context: {context}
Instruction: {user_query}<|eot_id|><|start_header_id|>assistant<|end_header_id|>

"""
```

## ДџЕёВ§В© 15. INFERENCE

```python
from unsloth import FastLanguageModel
FastLanguageModel.for_inference(model)

def ask(query):
    prompt = build_prompt(query)

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    outputs = model.generate(
        **inputs,
        max_new_tokens=400,
        temperature=0.7,
        eos_token_id=tokenizer.eos_token_id
    )

    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return decoded.split("### Response:")[-1].strip()

print(ask("Create a login screen with Jetpack Compose"))
```
